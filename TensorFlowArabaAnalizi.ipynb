```python
# drop : ile istenilen satır veya kolonu atar.
dataFrame = dataFrame.drop("transmission",axis=1)
```

```python
y=dataFrame["price"].values
x=dataFrame.drop("price",axis=1).values
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=10)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
sclear=MinMaxScaler()
```

```python
x_train=sclear.fit_transform(x_train)
x_test=sclear.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x_train.shape
```

```python
model = Sequential()
#dense: katman oluşturma
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))

model.add(Dense(1))# çıkış fonksiyonu

model.compile(optimizer="adam",loss="mse")
```

```python
model.fit(x=x_train, y=y_train,validation_data=(x_test,y_test), batch_size=250, epochs=300)
```

```python
kayipData=pd.DataFrame(model.history.history)
```

```python
kayipData.head()
```

```python
kayipData.plot()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
tahminDizisi=model.predict(x_test)
```

```python
tahminDizisi
```

```python
mean_absolute_error(y_test, tahminDizisi)
```
